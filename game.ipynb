```python
# Calculator
while True:
    num1 = float(input("Enter first number:"))
    num2 = float(input("Enter another number:"))
    print("You entered:", num1 ,",", num2 )

    operator = input("Enter an operation (+,-,*,/,**):")
    print("You chose:",operator)

    while operator not in ["+","-","*","/",]:
        print("Please choose a valid operator again.")
        operator = input("Enter an operation (+,-,*,/,**):")

    if operator == "+":
        result = num1 + num2
        print(f"{num1} + {num2} = {result}")

    elif operator == "-":
        result = num1 - num2
        print(f"{num1} - {num2} = {result}")

    elif operator == "*":
        result = num1 * num2
        print(f"{num1} * {num2} = {result}")

    elif operator == "**":
        result = num1 ** num2
        print(f"{num1} ** {num2} = {result}")

    elif  operator == "/":
        if num2 != 0:
            result = num1 / num2
            print(f"{num1} /  {num2} = {result}")
        elif num2 == 0:
            print("Number cannot be divisible by zero.")
    else:
        print("Invalid Calculation.")
    reset = input("Press 'C' to calcualte again. And press any other key to exist.")
    if reset.lower() != "c":
        break
print("GOODBYE!")
```

```python
# grade calculator
while True:
    print("Grade Calculator")

    subject_count = int(input("Enter total subjects you have:"))
    print(f"You have {subject_count} subjects.")
    
    while subject_count <= 0:
        print("Invalid subjects. Please enter your subject again.")
        subject_count = int(input("Enter total subjects you have:"))

    total_score = 0

    for i in range(subject_count):
        score = int(input(f"Enter you score in subject {i + 1} is: "))
        total_score = total_score + score
        while score <0 or score > 100:
            print("Invalid score. Please enter your score again.")
            score = input(f"Enter you score in subject {i + 1} is: ")

    average_score = total_score / subject_count
    if average_score >= 90 :
        grade = "A"
    elif average_score >= 80 :
        grade = "B"
    elif average_score >= 70:
        grade = "C"
    elif average_score >= 60:
        grade = "D"
    else:
        grade ="F"
    print()

    print("--- Student Report ---")
    print("Your total subject is:", subject_count)
    print("Your total score is:", total_score)
    print("Your average score is:", average_score)
    print("Your grade is:", grade)
    print()
    reset = input("Press 'C' to do another calculation. OR press anyother key to exist.")
    if reset.lower() != "c":
        break
```

```python
# ------------------ Credentials ------------------
usernameb = "boss"
passwordb = "boss"

usernamem = "manage"
passwordm = "manage"

usernamee = "employee"
passworde = "employee"

# ------------------ File Handling Helpers ------------------
def log_attempt(role, username, success):
    """Log login attempts to a file"""
    with open("login_log.txt", "a") as file:
        status = "Success" if success else "Failed"
        file.write(f"{role} | {username} | {status}\n")

def save_reports():
    """Save reports to a file (Boss only)"""
    reports = ["Sales Report Q1", "Employee Performance Report", "Budget Report"]
    with open("reports.txt", "w") as file:
        for report in reports:
            file.write(report + "\n")

def view_reports():
    """Read reports from file"""
    try:
        with open("reports.txt", "r") as file:
            print("Reports:")
            for line in file:
                print("-", line.strip())
    except FileNotFoundError:
        print("No reports found. Please save reports first.")

def log_action(role, action):
    """Log menu actions to a file"""
    with open("actions_log.txt", "a") as file:
        file.write(f"{role} performed: {action}\n")

# ------------------ Menus ------------------
def login():
    print("-------Login System---------")
    print("1. Boss")
    print("2. Manager")
    print("3. Employee")
    print("4. Exit")
    print("-" * 25)

def show_menu(role):
    while True:
        if role == "Boss":
            print("\n--- Boss Menu ---")
            print("1. View all reports")
            print("2. Manage employees")
            print("3. Approve budgets")
            print("4. Logout")

            choice = int(input("Please enter your choice: "))
            if choice == 1:
                save_reports()
                view_reports()
                log_action("Boss", "View all reports")
            elif choice == 2:
                print("Managing employees... ")
                log_action("Boss", "Manage employees")
            elif choice == 3:
                print("Approving budgets... ")
                log_action("Boss", "Approve budgets")
            elif choice == 4:
                print("Logging out......")
                break
            else:
                print("Invalid choice. Try Again!")

        elif role == "Manager":
            print("\n--- Manager Menu ---")
            print("1. Assign tasks")
            print("2. Review team progress")
            print("3. Log out")

            choice = int(input("Please enter your choice: "))
            if choice == 1:
                print("Assigning tasks... ")
                log_action("Manager", "Assign tasks")
            elif choice == 2:
                print("Reviewing team progress... ")
                log_action("Manager", "Review team progress")
            elif choice == 3:
                print("Logging out.....")
                break
            else:
                print("Invalid choice. Try Again!")

        elif role == "Employee":
            print("\n--- Employee Menu ---")
            print("1. View assigned tasks")
            print("2. Update task status")
            print("3. Logout")

            choice = int(input("Please enter your choice: "))
            if choice == 1:
                print("Viewing assigned tasks... ")
                log_action("Employee", "View assigned tasks")
            elif choice == 2:
                print("Updating task status... ")
                log_action("Employee", "Update task status")
            elif choice == 3:
                print("Logging out......")
                break
            else:
                print("Invalid choice. Try Again!")

# ------------------ Login System ------------------
def login_system():
    choice = int(input("Choose between (1-4): "))
    attempts = 0
    print()

    while choice not in [1, 2, 3, 4]:
        print("Please choose number between 1 to 4 only!")
        choice = int(input("Choose between (1-4): "))

    while True:
        if choice == 1:
            print("You chose Boss.")
            print("-" * 20)

            while attempts < 3:
                username1 = input("Enter your username: ")
                password1 = input("Enter your password: ")

                if username1 == usernameb and password1 == passwordb:
                    print("Login Successful. Welcome to the System.")
                    log_attempt("Boss", username1, True)
                    show_menu("Boss")
                    break
                else:
                    attempts += 1
                    log_attempt("Boss", username1, False)
                    print(f"Invalid Password or Username. Attempts Left: {3 - attempts}")
                    print("-" * 20)

                    if attempts == 3:
                        print("Too many attempts failed. Account locked temporarily.")

            break

        elif choice == 2:
            print("You chose Manager.")
            print("-" * 20)

            while attempts < 3:
                username2 = input("Enter your username: ")
                password2 = input("Enter your password: ")

                if username2 == usernamem and password2 == passwordm:
                    print("Login Successful. Welcome to the System.")
                    log_attempt("Manager", username2, True)
                    show_menu("Manager")
                    break
                else:
                    attempts += 1
                    log_attempt("Manager", username2, False)
                    print(f"Invalid Password or Username. Attempts Left: {3 - attempts}")
                    print("-" * 20)

                    if attempts == 3:
                        print("Too many attempts failed. Account locked temporarily.")

            break

        elif choice == 3:
            print("You chose Employee.")
            print("-" * 20)

            while attempts < 3:
                username3 = input("Enter your username: ")
                password3 = input("Enter your password: ")

                if username3 == usernamee and password3 == passworde:
                    print("Login Successful. Welcome to the System.")
                    log_attempt("Employee", username3, True)
                    show_menu("Employee")
                    break
                else:
                    attempts += 1
                    log_attempt("Employee", username3, False)
                    print(f"Invalid Password or Username. Attempts Left: {3 - attempts}")
                    print("-" * 20)

                    if attempts == 3:
                        print("Too many attempts failed. Account locked temporarily.")

            break

        else:
            print("You have successfully logged out!")
            break

# ------------------ Program Loop ------------------
while True:
    login()
    login_system()
    again = input("Do you want to login again? (y/n): ")
    if again.lower() != "y":
        print("Goodbye!")
        break
```
