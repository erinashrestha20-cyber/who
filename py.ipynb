```python
def calculate_total_after_discount(total, discount_percent):
    if total > 100:
        excess = total - 100  
        discount = (discount_percent / 100) * excess  
        total -= discount 
    
    return total

# For the first customer
total = float(input("Total? "))
discount_percent = float(input("Discount percentage? "))
total = calculate_total_after_discount(total, discount_percent)
print(f"The final amount to pay is: ${total}")
print()
print("Next Customer Please")

# For the second customer
total = float(input("Total? "))
discount_percent = float(input("Discount percentage? "))
total = calculate_total_after_discount(total, discount_percent)
print(f"The final amount to pay is: ${total}")
```

```python

def processString(s):
    length = len(s)
    uppercase_s = s.upper()
    vowels = "AEIOU"
    vowel_count = sum(1 for char in uppercase_s if char in vowels)
    return length, uppercase_s, vowel_count

text = "Welcome to python Programming!"
length, uppercase_text, vowel_count = processString (text)
processString(text)    
```

```python
def manage_friends():
    file_name = "friends.txt"

    # Create/clear the file at the start
    with open(file_name, "w") as file:
        pass  

    friends = []

    while True:
        name = input("Enter a name (or type 'DONE' to finish): ").strip()

        if name.upper() == "DONE":
            break

        # Append the new name to the file
        with open(file_name, "a") as file:
            file.write(name + "\n")

        # Read all names from the file into the list
        with open(file_name, "r") as file:
            friends = file.read().splitlines()

        # Sort and display the list
        friends.sort()
        print("Friends list (sorted):", friends)


# Run the program
manage_friends()
```

```python
# Program to search product price from stationery.txt

# Open the file in read mode
with open("stationery.txt","w") as file:
    file.write("pencil,2.10\n" )
    file.write("eraser,1.10\n")
    
with open("stationery.txt", "r") as file:
    # Read all lines and store in a dictionary
    prices = {} # all the written lines are stored in variable prices as dictionary
    for line in file:  ## reads each line
        item, price = line.strip().split(",") ##strip() removes unnecessary space from each line
        #while splilt(), separates two character using comma  
        prices[item.lower()] = price  ## this converts enterd item into lower letter

# Ask user for product to search
product = input("Enter product to search : ").lower()  ## user is asked to enter the item he wants to search for

# Check if product exists in dictionary
if product in prices:
    print("Price is : RM", prices[product])
else:
    print("Product not found.")

print("\nEnd of program")
```

```python
def calculate_fine():
    delayed_day = int(input("Enter number of days:"))
    print(f"Enter the Number of days: {delayed_day}")

    while True:
        if delayed_day == 0:
            print("N0 fine.")
        elif delayed_day > 0 and delayed_day <=5:
            fine = delayed_day * 0.5
            print("Fine Amount Pay to RM:", fine)
        elif delayed_day > 5 and delayed_day <=10:
            fine = (5 * 0.5) + ((delayed_day - 5) * 1)
            print("Fine Amount Pay to RM:", fine)
        elif delayed_day > 10 and delayed_day <= 30:
            fine = (5 * 0.5) + (5 * 1) + ((delayed_day - 10) * 5)
            print("Fine Amount Pay to RM:", fine)
        else:
            print("Your membership is cancelled since you returned book after 30 days.")
        break
    
calculate_fine()
```
